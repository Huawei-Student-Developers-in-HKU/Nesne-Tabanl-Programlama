İşte ana sınıf(ParentClass) ve alt sınıf arasında bağ oluşturma bu şekilde gerçekleşiyor. Biraz daha açacak olursak.

````python
class Vehicle:
    def __init__(self, make, color, model):
        self.make = make
        self.color = color
        self.model = model

    def printDetails(self):
        print("Manufacturer:", self.make)
        print("Color:", self.color)
        print("Model:", self.model)


class Car(Vehicle):
    def __init__(self, make, color, model, doors):
        # Ana sınıf ile eşitleme
        Vehicle.__init__(self, make, color, model)
        # Sınıfa ait değişkeni eşitleme
        self.doors = doors

    def printCarDetails(self):
        self.printDetails()
        print("Doors:", self.doors)


obj1 = Car("Suzuki", "Grey", "2015", 4)
obj1.printCarDetails()
````

Araç adında bir ana sınıf oluşturduk ve her araçta ortak olan özellikleri de consturcter methodun ile belirttik. Daha sonra bu değişkenleri yazdıracak bir method oluşturduk. Buraya kadar herşey ana sınıf içindi.

Şimdi ana sınıfa bağlı bir alt sınıf açtık, araölar sınıfına bağlı bir araba sınıfı. Yeni sınıfında tıpkı ana sınıfta olan daha öncede dediğim gibi her araçta orak olan özellikleri var ama ekstra olarak bir de kapı sayısı değişkenini consturcter method ile belirttik.

Daha sonra ise araba sınıfınde yeni bir obje oluşturukduğunda bu obje için girilen ve ana sınıfta da bulunan değişkenleri ana sınıfa da eklememiz gerekir. Bunun içinde ana sınıftaki constructer methodunu çağırdık ve bu değişkenleri eşitlerdik. Daha sonra geriye kalan kapı değişkenini sadece bu sınıfa ait olduğı için self ile eşitledik.

# Super Function
super() fonksiyonu alt sınıftan bir üst sınıfı temsil etmesi için kullanılan bir yöntemdir. Kısa yol olarak düşünebilirsiniz. 

````python
class Vehicle: 
    fuelCap = 90

class Car(Vehicle):  
    fuelCap = 50

    def display(self):
        print("Fuel cap from the Vehicle Class:", super().fuelCap)
        print("Fuel cap from the Car Class:", self.fuelCap)


obj1 = Car()  
obj1.display()
````

**Calling A Method**

````python
class Vehicle:  
    def display(self):  
        print("I am from the Vehicle Class")


class Car(Vehicle): 
  
    def display(self):
        super().display()
        print("I am from the Car Class")


obj1 = Car()  
obj1.display()
````

**Using With Initiliazers**

````python
class ParentClass():
    def __init__(self, a, b):
        self.a = a
        self.b = b


class ChildClass(ParentClass):
    def __init__(self, a, b, c):
        super().__init__(a, b)
        self.c = c


obj = ChildClass(1, 2, 3)
print(obj.a)
print(obj.b)
print(obj.c)
````

# Initiliaze Objects 

Bir sınıfın nesnesini başlatmak için kullanılır, sınıf yapısı çağırıldığında arka tarafta kendi kendine çalışan bir fonksiyondur aslında. Diğer dillerdeki constructer methodlara benzese de tam olarak aynısı değildir. 

Initiliaze methodu özel bir methoddur ve `__init__` şeklindedir. Fonksiyonu ilk parametresi `self`'dir ama bu kısmada sonra geleceğiz.

```python
class employee:
  def __init__(self, ID, salary, department):
    self.ID = ID
    self.salary = salary
    self.department = department
    print("Sınıf başladı")
#####################################    

Newt = employee(986, 7000, "Security")
print("ID :: ", Newt.ID)
print("Salary :: ", Newt.salary)
print("Department :: ",Newt.department)

John = employee(856, 8000, "Transporting")
print("ID :: ", John.ID)
print("Salary :: ", John.salary)
print("Department :: ",John.department)
```

Tekrar etmek gerekirse initiliazer methodları sınıf çağırıldığında başlayan, yani `Newt = employee()` anında çalışmaya başlayan bir fonksiyondur. Doğal olarak da bir fonksiyona ihtiyacı olduğu parametreleri belirtmeniz gerekir bizimde yaptığımız gibi :: `employee(986, 7000, "Security")`

Ama bu parametreleri opsiyonel hale de getirebiliriz bir diğer deyişle bu parametreleri isteğe bağlı girilecek şekilde düzenleyebilirz. Bunun için de fonksiyonun parametrelerine default bir değer atamalıyız ki eğer kullanıcı değer girmez ise default değer işleme sokulsun.



```python
class employee:
  def __init__(self, ID = None, salary = 0, department = None):
    self.ID = ID
    self.salary = salary
    self.department = department

Newt = employee(7000, "Security")
print("ID :: ", Newt.ID)
print("Salary :: ", Newt.salary)
print("Department :: ",Newt.department)
```

Dikkat etmeniz gereken şey ise python kendi yapısı gereği girdiğiniz parametreleri sıra ile yerleştirir. Yani yukarıdaki `durumda ID = 7000 `,`salary = "Security"` ve `department = None` şeklinde atandı. Bu işlemin düzgün yapılışı ise :

```python
Newt = employee(salary = 7000, department ="Security")
print("ID :: ", Newt.ID)
print("Salary :: ", Newt.salary)
print("Department :: ",Newt.department)
```

# Class and Instance  Değişkenleri

Class değişkenleri sınıfta bulunan tüm objeler için aynı olan ve herhangi bir obje için değiştirilme durumunda tğm objeler için değişen değişkenlerdir.

Instance değişkenleri için her objeye özel, objeden objeye değişiklik gösteren, bir objede değiştirilmesi durumunda bir tek o obje için değişen değişkenlerdir.

```python
class Player:
  # Class Variable
  teamname = "LiverPool"

  def __init__(self, name):
    # Instance Variable
    self.name = name

p1 = Player("John")
p2 = Player("Newt")

print("Team Name for p1 is :: ", p1.teamname)
print("Team Name for p2 is :: ", p2.teamname)

print("Player name for p1 is :: ", p1.name)
print("Player name for p2 is :: ", p2.name)
```

# Sınıflara Method Ekleme


```python
class Employee:
    def __init__(self, ID=None, salary=None, department=None):
        self.ID = ID
        self.salary = salary
        self.department = department

    def tax(self):
        return (self.salary * 0.2)

    def salaryPerDay(self):
        return (self.salary / 30)


Steve = Employee(3789, 2500, "Human Resources")

# Steveée ait değerler
print("ID =", Steve.ID)
print("Salary", Steve.salary)
print("Department:", Steve.department)

print("Tax paid by Steve:", Steve.tax())
print("Salary per day of Steve", Steve.salaryPerDay())
```

# Bir Methodu Aşırı Yükleme
Aynı ada ancak farklı imzalara (veya bağımsız değişkenlere) sahip iki methodumuz olduğunda aşırı yükleme gerçekleşir. Diğer programlam dillerinde bir sınıfta aynı isimle iki veya daha fazla metot uygulayabilirken ne yazık ki Python programla dilinde böyle bir şey mümkün değildir. Aynı isimli methodlar arasında en son yazılanı geçerli sayar. Bu yüzden daha önce initiliaze methodun da yaptığımız gibi kullaılma ihtimali olan parametrelri ve onların default değerini tanımlıyoruz.

```python
class Employee:
    # defining the properties and assigning them None to the
    def __init__(self, ID=None, salary=None, department=None):
        self.ID = ID
        self.salary = salary
        self.department = department

    # method overloading
    def demo(self, a, b, c, d=5, e=None):
        print("a =", a)
        print("b =", b)
        print("c =", c)
        print("d =", d)
        print("e =", e)
      

# cerating an object of the Employee class
Steve = Employee()

# Printing properties of Steve
print("Demo 1")
Steve.demo(1, 2, 3)
print("\n")

print("Demo 2")
Steve.demo(1, 2, 3, 4)
print("\n")

print("Demo 3")
Steve.demo(1, 2, 3, 4, 5)
```

**Overload Method Kullanımının Faydaları**
* Hafızada yeni ethod için alan oluşturmaya gerek kalmaz.
* Kodu daha okunabilir yapar.
* İşlem hızını arttırır.
* Polimorfizm işlemine uygun hale getirir. 

# Sınıf Yapılarındaki Fonksiyon Türleri
Bir sınıfa ekleyebileceğiniz 3 tip method vardır. Bunlar : 

*   instance methods
*   class methods
*   static methods



Daha önceki kısımlarda kullandığımız method instance methodu idi. Adın da da anlaşılabileceği gibi instance objeler ile yapılan işlemlerde kullanılan method yani def `method_adı(self):` şeklinde tanımlanan methodlardır. Buradaki self'in instance değişkenleri temsil ettiğini zaten biliyoruz.

```python
class MyClass:
    def __init__(self):
      self.instanceVariable = "Hello World"

    def demo(self):
        return self.instanceVariable
```

Class methodu ise instace methoduna çok benzer. Fark ise class methodu class değişkenleri ile işlem yapar. Daha önce bahsettiğimiz her obje için sabit olan değişkenlerden bahsediyorum. Class methodunu tanımlamak için ise `@classmethod `ve `cls` kullanılır. Buradaki cls self gibi işlev görürü ve kullanıcının isteğine göre değişebilir ancak genel kullanım cns'dir.




```python
class MyClass:
    classVariable = "Hello World"

    @classmethod
    def demo(cls):
        return cls.classVariable
```

Son olarak ise static methodlar vardır. Bunlar sınıfa bağlı ama herhangi bir obje ile doğrudan bağı olmayan methodlardır. Yani instance methodlar instance değişkenler ile, class methodları class methodları ile doğrudan bağlı iken static methodlar sıradan bir bağı olmayan methodlardır. 

```python
class MyClass:

    @staticmethod
    def demo():
        print("I am a static method")
```

# Access Modifiers

Python'da farklı veri üyelerine ve üye işlevlerine erişim kısıtlamaları getirebiliriz. Bir değişken veya method ya public yada private olabilir. Herhangi bir işlem yapmadan, düz şekilde yani şuana kadar tanımladığımız değişkenler ve methodlar public'tir. Python default olarak bunları public yapar.

Eğer bir değişnein veya methodun private olmasını istiyorsanız isminden önce iki alt çizgi(__) getirin tıpkı aşağıda olduğu gibi.

```python
class Employee:
    def __init__(self, ID, salary):
        self.ID = ID
        self.__salary = salary  

Steve = Employee(3789, 2500)
print("ID:", Steve.ID)
print("Salary:", Steve.__salary) # HATA !!!
```

Yukarıda da görüldüğü gibi __salary değişkenine sınıf dışından erişmek artık mümkğn değil. Gördüğünüz gibi hata Employee sınıfında böyle bir obje olmadığını söylüyor. Aşağıda da bir private method örneği var.

```python
class Employee:
    def __init__(self, ID, salary):
        self.ID = ID
        self.__salary = salary  

    # Normal Method
    def displaySalary(self): 
        print("Salary:", self.__salary)

    # Private Method
    def __displayID(self):  
        print("ID:", self.ID)


Steve = Employee(3789, 2500)
Steve.displaySalary
Steve.__displayID()  # ! HATA
```

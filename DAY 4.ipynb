# Aggregation
Aggregaton Has-A modelini takip eder. Bu, bir sınıfın diğerinin nesnesine sahip olduğu iki sınıf arasında bir ebeveyn-çocuk ilişkisi yaratır.



```python
class Country:
    def __init__(self, name=None, population=0):
        self.name = name
        self.population = population

    def printDetails(self):
        print("Country Name:", self.name)
        print("Country Population", self.population)


class Person:
    def __init__(self, name, country):
        self.name = name
        self.country = country

    def printDetails(self):
        print("Person Name:", self.name)
        self.country.printDetails()


c = Country("Wales", 1500)
p = Person("Joe", c)
p.printDetails()

# deletes the object p
del p
print("")
c.printDetails()
```

# Composition
Kompozisyon, sınıfınızdaki diğer sınıf nesnelerine erişme uygulamasıdır. Böyle bir senaryoda, diğer sınıfın nesnesini oluşturan sınıf, sahibi olarak bilinir ve o nesnenin ömründen sorumludur.

Kompozisyon ilişkileri, parçanın tüm nesnenin bir parçasını oluşturması gereken ilişkilerin Parçasıdır. Karmaşık bir birim yapmak için diğer sınıfların daha küçük parçalarını ekleyerek kompozisyon elde edebiliriz.

```python
class Engine:
    def __init__(self, capacity=0):
        self.capacity = capacity

    def printDetails(self):
        print("Engine Details:", self.capacity)


class Tires:
    def __init__(self, tires=0):
        self.tires = tires

    def printDetails(self):
        print("Number of tires:", self.tires)


class Doors:
    def __init__(self, doors=0):
        self.doors = doors

    def printDetails(self):
        print("Number of doors:", self.doors)


class Car:
    def __init__(self, eng, tr, dr, color):
        self.eObj = Engine(eng)
        self.tObj = Tires(tr)
        self.dObj = Doors(dr)
        self.color = color

    def printDetails(self):
        self.eObj.printDetails()
        self.tObj.printDetails()
        self.dObj.printDetails()
        print("Car color:", self.color)


car = Car(1600, 4, 2, "Grey")
car.printDetails()
```
